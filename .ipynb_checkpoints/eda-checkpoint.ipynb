```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 500)
```

```python
url ='https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
can = df[df['iso_code']=='CAN']
#can.to_csv('data/canada.csv')

plt.figure(figsize=(14,8))
plt.plot(can['date'], can['total_cases'], label='total cases')
plt.plot(can['date'], can['total_tests'], label='total tests')
plt.legend(loc='best')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.title('Canada')
plt.show()
```
